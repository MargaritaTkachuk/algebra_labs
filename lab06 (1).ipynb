### 1. Розширення полів

1. Побудуйте розширення $K=\mathbb{Q}(\sqrt[3]{2},i)$ поля $\mathbb{Q}$.
2. Знайдіть степінь розширення $[K:\mathbb{Q}]$ та базис $K$ над $\mathbb{Q}$.
3. Задайте елемент $a=\sqrt[3]{2}+i$ і знайдіть його мінімальний многочлен $m_a(x)$ над полем $\mathbb{Q}$.
4. Побудуйте розширення $L=\mathbb{Q}(\sqrt[3]{2}+i)$ як факторкільце $\mathbb{Q}[x]/(m_a(x))$.
5. Перевірте, чи є поля $K$ та $L$ рівними, порівнявши степені розширення.



```python
# Поліноми над QQ
R.<x> = QQ[]

# 1. Побудова K = Q(sqrt(2), i)

# 1) Перше розширення: Q( sqrt(2) )
K1.<s2> = NumberField(x^2 - 2)
print("K1 =", K1)

# 2) Друге розширення: додаємо i
S.<y> = PolynomialRing(K1)
K.<ii> = K1.extension(y^2 + 1)

print("\nK = Q(sqrt(2), i) =", K)


# 2. Ступінь і базис
print("K = Q(sqrt(2), i)")
print("Absolute degree:", K.absolute_degree())
print("\n   Базис K над Q (абсолютний):")
# Отримуємо базис через векторний простір
V, from_V, to_V = K.absolute_vector_space()
basis_K = [from_V(e) for e in V.basis()]
for idx, b in enumerate(basis_K):
    print(f"   b{idx+1} = {b}")
print()


# 3. Мінімальний поліном a = sqrt(2) + i
a = s2 + ii
ma = a.minpoly()
print("minpoly =", ma)

# 4. Поле L = Q(a)
L = QQ.extension(ma, 'alpha')
alpha = L.gen()
print("Degree[L:Q] =", L.absolute_degree())

# 5. Спроба вбудувати L → K
phi = L.hom([a], K)

if phi.is_injective():
    print("L embeds into K ⇒ L ⊆ K")
else:
    print("Embedding failed")

```

### 2. Поля розкладу

1. Побудуйте поле розкладу $L$ многочлена $f(x)=x^4+x^2+1$ над полем $\mathbb{Q}$.
2. Знайдіть степінь та базис $L$ над $\mathbb{Q}$.
3. Розкладіть многочлен $f(x)$ на лінійні множники над полем $L$.
4. Побудуйте поле розкладу $K$ многочлена $g(x)=x^5+3x^2+2$ над полем $\mathbb{Z}_3$.
5. Знайдіть степінь та базис $K$ над $\mathbb{Z}_3$.
6. Розкладіть многочлен $g(x)$ на лінійні множники над полем $K$.

```python
###############################################################
# 1. Побудуйте поле розкладу L многочлена f(x) = x^4 + x^2 + 1
###############################################################

R.<x> = QQ[]
f = x^4 + x^2 + 1

L = f.splitting_field('x')
alpha = L.gen()

print("1. Многочлен f(x):", f)
print("Поле розкладу L:", L)
print("-"*40)


###############################################################
# 2. Степінь та базис поля L над Q
###############################################################

deg_L = L.absolute_degree()

print("Базис L над Q:")
# Отримуємо базис через векторний простір
V, from_V, to_V = L.absolute_vector_space()
basis_L = [from_V(e) for e in V.basis()]
for idx, b in enumerate(basis_L):
    print(f"   b{idx+1} = {b}")
print()

print("2. Степінь [L:Q] =", deg_L)

print("-"*40)


###############################################################
# 3. Розклад f(x) на лінійні множники в L
###############################################################

RL.<t> = PolynomialRing(L)
f_L = RL(f)

print("3. Розклад f(x) над L:")
print(f_L.factor())
print("-"*40)


###############################################################
# 4. Побудова поля розкладу g(x) = x^5 + 3x^2 + 2 над F3
###############################################################

S.<y> = GF(3)[]
g = y^5 + 3*y^2 + 2   # У F3 це y^5 + 2

K = g.splitting_field('x')
beta = K.gen()

print("4. Многочлен g(x) над F3:", g)
print("Поле розкладу K =", K)
print("-"*40)


###############################################################
# 5. Степінь та базис поля K над F3
###############################################################

deg_K = K.degree()
print("5. Степінь [K:F3] =", deg_K)
print("Базис K над F3:")
prim = K.multiplicative_generator()
print(f"   Примітивний елемент поля: β = {prim}")
print(f"   Порядок β: {prim.multiplicative_order()}")
print("   Базис {1, β, β^2, ..., β^(n-1)}:")
for i in range(deg_K):
    print(f"   β^{i} = {prim^i}")   # У скінченних полях basis() працює


print("-"*40)



###############################################################
# 6. Розклад g(x) над K
###############################################################

SK.<z> = PolynomialRing(K)
g_K = SK(g)

print("6. Розклад g(x) над K:")
print(g_K.factor())
print("-"*40)
```

### 3. Скінченні поля

1. Побудуйте поле $F$ з $8$ елементів. Виведіть таблицю множення.
2. Побудуйте факторкільця $K=\mathbb{Z}_2[x]/(x^3+x+1)$ та $L=\mathbb{Z}_2[x]/(x^3+x^2+1)$.
3. Перевірте, що $K$ та $L$ є полями з $8$ елементів.
4. Знайдіть первісні елементи для полів $K$ та $L$.
5. Встановіть ізоморфізм між полями $K$ та $L$.


```python
print("=" * 70)
print("РОБОТА З ПОЛЯМИ З 8 ЕЛЕМЕНТІВ")
print("=" * 70)

###############################################################
# 1. Побудова поля F з 8 елементів та таблиці множення
###############################################################

print("\n" + "=" * 70)
print("1. ПОБУДОВА ПОЛЯ F З 8 ЕЛЕМЕНТІВ")
print("=" * 70)

# Побудова поля GF(8) = GF(2^3)
F.<a> = GF(8, modulus='primitive')  # 'a' - примітивний елемент
print(f"Поле F = GF(8) = GF(2^3)")
print(f"Характеристика: {F.characteristic()}")
print(f"Порядок: |F| = {F.order()}")
print(f"Примітивний елемент: a = {a}")
print(f"Порядок елемента a: {a.multiplicative_order()}")

# Всі елементи поля
print(f"\nВсі елементи поля F:")
elements = list(F)
for i, elem in enumerate(elements):
    print(f"  {i}: {elem}")

# Таблиця множення
print("\nТаблиця множення поля F:")
print("   " + " ".join(f"{str(elem):>3}" for elem in elements))
print("  " + "-" * (4 * len(elements) + 1))

for i, row_elem in enumerate(elements):
    row = [str(row_elem)] + [str(row_elem * col_elem) for col_elem in elements]
    print(f"{str(row_elem):>2} | " + " ".join(f"{val:>3}" for val in row[1:]))

print("\nПримітивні елементи (генератори мультиплікативної групи):")
primitive_elements = [elem for elem in F if elem != 0 and elem.multiplicative_order() == 7]
for elem in primitive_elements:
    print(f"  {elem} (порядок: {elem.multiplicative_order()})")

###############################################################
# 2. Побудова факторкілець K та L
###############################################################

print("\n" + "=" * 70)
print("2. ПОБУДОВА ФАКТОРКІЛЕЦЬ K ТА L")
print("=" * 70)

# Кільце многочленів над Z2
Z2x.<x> = PolynomialRing(GF(2))

# Два незвідних многочлени ступеня 3 над Z2
f1 = x^3 + x + 1
f2 = x^3 + x^2 + 1

print(f"Многочлен f1(x) = {f1}")
print(f"Многочлен f2(x) = {f2}")

# Перевірка незвідності
print(f"\nПеревірка незвідності:")
print(f"  f1(x) незвідний над Z2: {f1.is_irreducible()}")
print(f"  f2(x) незвідний над Z2: {f2.is_irreducible()}")

# Побудова факторкілець
K = Z2x.quotient(f1, names=('alpha',))  # alpha - образ x у факторі
L = Z2x.quotient(f2, names=('beta',))   # beta - образ x у факторі

# Отримуємо генератори
alpha = K.gen()
beta = L.gen()

print(f"\nФакторкільце K = Z2[x]/({f1})")
print(f"  Клас α = x + ({f1}) = {alpha}")
print(f"  Елементи K: {list(K)}")

print(f"\nФакторкільце L = Z2[x]/({f2})")
print(f"  Клас β = x + ({f2}) = {beta}")
print(f"  Елементи L: {list(L)}")

###############################################################
# 3. Перевірка, що K та L є полями з 8 елементів
###############################################################

print("\n" + "=" * 70)
print("3. ПЕРЕВІРКА, ЩО K ТА L Є ПОЛЯМИ")
print("=" * 70)

print(f"\nДля K = Z2[x]/({f1}):")
print(f"  Кількість елементів: {K.cardinality()}")
print(f"  Характеристика: {K.characteristic()}")
print(f"  Є полем: {K in Fields()}")

print(f"\nДля L = Z2[x]/({f2}):")
print(f"  Кількість елементів: {L.cardinality()}")
print(f"  Характеристика: {L.characteristic()}")
print(f"  Є полем: {L in Fields()}")

# Перевірка, що α та β задовольняють свої рівняння
print(f"\nПеревірка рівнянь:")
print(f"  У K: α^3 + α + 1 = {alpha^3 + alpha + 1} = 0? {alpha^3 + alpha + 1 == 0}")
print(f"  У L: β^3 + β^2 + 1 = {beta^3 + beta^2 + 1} = 0? {beta^3 + beta^2 + 1 == 0}")

# Всі елементи K та L
print(f"\nВсі елементи K (8 елементів):")
for i, elem in enumerate(K):
    print(f"  {i}: {elem}")

print(f"\nВсі елементи L (8 елементів):")
for i, elem in enumerate(L):
    print(f"  {i}: {elem}")

###############################################################
# 4. Пошук первісних елементів
###############################################################
print("4. Пошук первісних елементів")
print("(Первісний елемент породжує мультиплікативну групу поля, його порядок = 7)")

def get_primitive_elements(Field, name):
    primitives = []
    group_order = Field.cardinality() - 1 # Порядок мультиплікативної групи (7)
    
    # Знаходимо прості множники порядку групи (для 7 це просто 7)
    prime_factors = [p for p, e in factor(group_order)]
    
    # Перебираємо всі елементи поля, крім нуля
    for elem in Field:
        if elem == 0: continue
        
        # Елемент g первісний, якщо g^(group_order) == 1 
        # і g^(group_order/p) != 1 для всіх простих дільників p.
        
        is_primitive = True
        
        # 1. Перевірка, чи елемент в правильному степені дає 1 (теорема Лагранжа гарантує це для поля, але про всяк випадок)
        if elem^group_order != 1:
            is_primitive = False
        else:
            # 2. Перевірка, чи не є порядок меншим
            for p in prime_factors:
                if elem^(group_order // p) == 1:
                    is_primitive = False
                    break
        
        if is_primitive:
            primitives.append(elem)
            
    print(f"\nПервісні елементи поля {name}:")
    print(primitives)
    return primitives

prim_K = get_primitive_elements(K, "K")
prim_L = get_primitive_elements(L, "L")

    
###############################################################
# 5. Встановлення ізоморфізму між K та L
###############################################################
print("--- 5. Встановлення ізоморфізму K -> L ---")

# Нам треба відобразити генератор 'k' поля K в якийсь елемент 'y' поля L.
# Генератор 'k' є коренем многочлена x^3 + x + 1 (бо K побудовано за цим модулем).
# Отже, його образ 'y' в L теж має бути коренем рівняння y^3 + y + 1 = 0.

target_element = None

print("Шукаємо елемент y в L, який задовольняє рівняння y^3 + y + 1 = 0...")

for elem in L:
    # Підставляємо елемент з L у поліном, який породжує K
    val = elem^3 + elem + 1
    if val == 0:
        target_element = elem
        print(f" -> Знайдено корінь: y = {target_element}")
        # Зазвичай ми хочемо відобразити первісний в первісний, 
        # але будь-який корінь підійде для гомоморфізму алгебр.
        break

if target_element is not None:
    print(f"\nБудуємо ізоморфізм phi: K -> L, де phi(k) = {target_element}")
    phi = K.hom([target_element], L)
    
    print("Ізоморфізм створено успішно.")
    
    print("\n--- Перевірка ізоморфізму ---")
    k = K.gen()

    elem_in_K = k^2 + k
    mapped_elem = phi(elem_in_K)
    
    # Обчислюємо очікуване значення вручну в L, замінюючи k на target_element
    expected_in_L = target_element^2 + target_element
    
    print(f"Елемент з K:     {elem_in_K}")
    print(f"Його образ в L:  {mapped_elem}")
    print(f"Очікуваний образ:{expected_in_L}")
    
    if mapped_elem == expected_in_L:
        print("Перевірка пройшла успішно: відображення коректне.")
else:
    print("Помилка: Не вдалося знайти відповідний елемент для ізоморфізму.")

```
