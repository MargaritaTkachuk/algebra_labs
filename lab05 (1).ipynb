### 2. Дільники нуля в кільцях

1. Задайте кільця $\mathbb{Z}_{30}$, $\mathbb{Z}_{12}\times\mathbb{Z}_{15}$, $M_3(\mathbb{Z}_3)$, $\mathbb{Z}_6[x]$.
2. Знайдіть кількість дільники нуля в цих кільцях.
3. Знайдіть мультиплікативну групу та її порядок. 



```python
# 1. Задати кільця
Z30 = Integers(30)
Z12xZ15 = cartesian_product([Integers(12), Integers(15)])
M3Z3 = MatrixSpace(Integers(3), 3)
Z6x = PolynomialRing(Integers(6), 'x')


# -------------------------------------------------------
# 2. Функції для знаходження дільників нуля
# -------------------------------------------------------

# Дільники нуля в Zn
def zero_divisors_Zn(n):
    Z = Integers(n)
    return [a for a in Z if a != 0 and gcd(a,n) != 1]

# Дільники нуля в Z12 × Z15
def zero_divisors_product():
    Z12 = Integers(12)
    Z15 = Integers(15)
    return [(a,b) for a in Z12 for b in Z15
            if (a,b) != (0,0) and (a == 0 or b == 0 or gcd(a,12)!=1 or gcd(b,15)!=1)]

# Дільники нуля у M3(Z3): матриці, що НЕ є оборотними, але не є нульовою
def zero_divisors_matrices():
    R = M3Z3
    return [A for A in R if A != 0 and A.det() == 0]

# У Z6[x] нескінченно багато дільників нуля → дамо приклад
def example_zero_divisors_Z6x():
    R = Z6x
    x = R.gen()
    return (2*x, 3*x)   # 2*x * 3*x = 6*x^2 = 0


# -------------------------------------------------------
# 3. Мультиплікативні групи кілець
# -------------------------------------------------------

# Мультиплікативна група Zn — це U(n)
def multiplicative_group_Zn(n):
    return Integers(n).unit_group()

# Для Z12 × Z15 група — добуток U(12) × U(15)
def multiplicative_group_product():
    return multiplicative_group_Zn(12), multiplicative_group_Zn(15)

# У M3(Z3) мультиплікативна група — GL(3,3)
def multiplicative_group_matrices():
    return GL(3, GF(3))

# У Z6[x] оборотні лише константи, що є одиницями Z6
def units_Z6x():
    R = Z6x
    return [1,5]   # бо gcd(a,6)=1


# -------------------------------------------------------
# Вивести результати
# -------------------------------------------------------

print("=== Zero divisors ===")
print("|Z_30 zero divisors| =", len(zero_divisors_Zn(30)))
print("|Z12×Z15 zero divisors| =", len(zero_divisors_product()))
print("|M3(Z3) zero divisors| =", len(zero_divisors_matrices()))
print("Example in Z6[x]:", example_zero_divisors_Z6x())

print("\n=== Multiplicative groups ===")
print("U(30): order =", multiplicative_group_Zn(30).order())
g12, g15 = multiplicative_group_product()
print("U(12): order =", g12.order())
print("U(15): order =", g15.order())
print("GL(3,3): order =", multiplicative_group_matrices().order())
print("Units in Z6[x]:", units_Z6x())
```

### 3. Корінь з одиниці за модулем $p$

1. Знайдіть перші сто простих чисел $p$, для яких факторкільце $\mathbb{Z}_p[x]/(x^2+1)$ є полем.  $\qquad$ % PolynomialRing(GF(p))
2. Знайдіть остачі при ділення цих простих чисел на $m$ для деяких значень $m=2,3,\ldots,10$.        
3. Спробуйте знайти закономірність і сформулювати гіпотезу про такі прості числа.



```python
def find_primes_for_field(num_primes):
    """
    Знаходить перші 'num_primes' простих чисел p, для яких x^2+1 незвідний над GF(p).
    """
    primes_list = []
    p = 2
    
    # Ітерація по простих числах
    while len(primes_list) < num_primes:
        # Створення поля GF(p) і кільця многочленів R
        Fp = GF(p)
        R = PolynomialRing(Fp, 'x')
        f = R([1, 0, 1]) # Многочлен x^2 + 1
        
        # Перевірка на незвідність
        if f.is_irreducible():
            primes_list.append(p)
        
        # Перехід до наступного простого числа
        p = next_prime(p)
        
    return primes_list

# Знайдемо перші 100 простих чисел, що задовольняють умову
first_100_primes = find_primes_for_field(100)
print(first_100_primes)

# first_100_primes - список, отриманий на попередньому кроці
m_values = range(2, 11)
remainders = {}

for m in m_values:
    # Обчислення остач для всіх 100 чисел
    rem_list = [p % m for p in first_100_primes]
    
    # Знахлдження унікальних остач
    unique_rem = sorted(list(set(rem_list)))
    remainders[m] = unique_rem

print(remainders)
```

Факторкільце Z\_p[ ]/\(x^2\+1\) є полем тоді й лише тоді, коли просте число p конгруентно 3 \(mod 4\)

### 4. Ідеали та факторкільця кільця многочленів

1. Задайте кільце многочленів $R=\mathbb{Z}_{7}[x]$. $\qquad$  %  R.\<x\>=Integers(7)[]
3. Задайте ідеали $I_1=(2x^3+5x+1)$, $I_2=(x^4+3x^3+2)$, $I_3=(x^5+x^3+x+1)$.  $\qquad$    %  I = R.ideal(f(x))
4. Побудуйте факторкільця $R/I$ і знайдіть його порядок. $\qquad$  %  R.quotient(І)
5. Перевірте, для яких ідеалів факторкільце $R/I$ є полем.
6. Перевірте, чи має елемент $(x^5+3x^2+2)+I$ обернений у факторкільці $R/I$ та знайдіть обернений.

```python
# Задання поля коефіцієнтів F = GF(7)
F = GF(7) 

# Задання кільця многочленів R = F[x]
R = PolynomialRing(F, 'x')
x = R.gen()

# Задання ідеалів I_1, I_2, I_3
f1 = 2*x^3 + 5*x + 1
f2 = x^4 + 3*x^3 + 2
f3 = x^5 + x^3 + x + 1

I1 = R.ideal(f1)
I2 = R.ideal(f2)
I3 = R.ideal(f3)

# print(f"Кільце R: {R}")
# print(f"Ідеал I1: {I1}")

S1 = R.quotient(I1, 'a')
S2 = R.quotient(I2, 'b')
S3 = R.quotient(I3, 'c')

# Знаходження порядку за допомогою вбудованої функції
order_S1 = S1.order()
order_S2 = S2.order()
order_S3 = S3.order()

print(f"Порядок |R/I1|: {order_S1} (7^3)")
print(f"Порядок |R/I2|: {order_S2} (7^4)")
print(f"Порядок |R/I3|: {order_S3} (7^5)")

# Перевірка незвідності
is_f1_irreducible = f1.is_irreducible()
is_f2_irreducible = f2.is_irreducible()
is_f3_irreducible = f3.is_irreducible()

print(f"f1 = {f1} незвідний (R/I є полем): {is_f1_irreducible}")
print(f"f2 = {f2} незвідний (R/I є полем): {is_f2_irreducible}")
print(f"f3 = {f3} незвідний (R/I є полем): {is_f3_irreducible}")


g = x^5 + 3*x^2 + 2

# Обчислення НСД
gcd_g_f1 = gcd(g, f1)

if gcd_g_f1.is_one():
    print(f"R/I1: НСД(g, f1) = 1. Обернений існує.")
    # Обчислення оберненого за допомогою розширеного алгоритму Евкліда (Bezout)
    # або безпосередньо у факторкільці.
    s1 = S1(g)^(-1)
    print(f"Обернений у R/I1: {s1}")
else:
    print(f"R/I1: НСД(g, f1) = {gcd_g_f1}. Обернений не існує.")
    

    
g = x^5 + 3*x^2 + 2

gcd_g_f2 = gcd(g, f2)
if gcd_g_f2.is_one():
    print(f"R/I2: НСД(g, f2) = 1. Обернений існує.")
    s2 = S2(g)^(-1)
    print(f"Обернений у R/I2: {s2}")
else:
    print(f"R/I2: НСД(g, f2) = {gcd_g_f2}. Обернений не існує.")
    
    
    
g = x^5 + 3*x^2 + 2

gcd_g_f3 = gcd(g, f3)
if gcd_g_f3.is_one():
    print(f"R/I3: НСД(g, f3) = 1. Обернений існує.")
    s3 = S3(g)^(-1)
    print(f"Обернений у R/I3: {s3}")
else:
    print(f"R/I3: НСД(g, f3) = {gcd_g_f3}. Обернений не існує.")
```
