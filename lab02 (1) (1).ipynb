```python
def subgroup_lattice_graph(G, name="Group"):
    subs = G.subgroups()
    
    G = Graph()
    
    for H in subs:
        G.add_vertex(H)
    
    for H in subs:
        for K in subs:
            if H.is_subgroup(K) and H != K:

                if not any(H.is_subgroup(M) and M.is_subgroup(K) and M not in (H, K) for M in subs):
                    G.add_edge(H, K)
    
    G.show(layout="spring", vertex_size=600, vertex_labels=False)
    return G
                
Z24 = CyclicPermutationGroup(24)
Q8  = QuaternionGroup()
D5  = DihedralGroup(5)
S4  = SymmetricGroup(4)

subgroup_lattice_graph(Z24, "Z24")
subgroup_lattice_graph(Q8, "Q8")
subgroup_lattice_graph(D5, "D5")
subgroup_lattice_graph(S4, "S4")

```

```python
Z24 = CyclicPermutationGroup(24)
Q8  = QuaternionGroup()
D5  = DihedralGroup(5)
S4  = SymmetricGroup(4)
```

```python
display(Z24.gens())
display(Q8.gens())
display(D5.gens())
display(S4.gens())
```

```python
from itertools import combinations

def minimal_generating_sets(G, max_size=3):
    elems = list(G)
    n = len(elems)
    result = []
    found_size = None
    
    for k in range(1, max_size + 1):
        for subset in combinations(elems, k):
            H = G.subgroup(subset)
            if H == G:
                if found_size is None:
                    found_size = k
                if k == found_size:
                    result.append(subset)
        if found_size is not None:
            break 
            
    return result

display(minimal_generating_sets(Z24))
display(minimal_generating_sets(Q8))
display(minimal_generating_sets(D5))
display(minimal_generating_sets(S4))
```
